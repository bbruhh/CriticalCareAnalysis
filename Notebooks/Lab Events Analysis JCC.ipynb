```python
import pandas as pd
```

```python
LABEVENTS = "../../Resources/LABEVENTS.csv"
D_LABITEMS = "../../Resources/D_LABITEMS.csv"
PATIENTS = "../../Resources/PATIENTS.csv"
```

```python
lab_df = pd.read_csv(LABEVENTS)

lab_df.head()
```

```python
items_df = pd.read_csv(D_LABITEMS)

items_df = items_df.drop(['ROW_ID'], axis=1)

items_df.head()
```

```python
joined_df = lab_df.merge(items_df, on='ITEMID')

joined_df.head()
```

```python
joined_df['FLUID'].value_counts()
```

```python
patient_df = pd.read_csv(PATIENTS)

patient_df = patient_df.drop(['ROW_ID'], axis=1)

patient_df.head()
```

```python
final_df = joined_df.merge(patient_df, on='SUBJECT_ID')

final_df.head()
```
